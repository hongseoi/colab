```python
#cumsum(누적합) 그래프
randn_cumsum = np.random.randn(1000).cumsum()
```

```python
fig, ax = plt.subplots(figsize=(10,6)) #공식 메뉴얼

ax.set_title("title")
ax.set_xlabel('stages')
ax.set_ylabel('random value')

ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize="small")
ax.annotate('250\n250.00\n', xy=(250,randn_cumsum[250]+0.5), xytext=(250, randn_cumsum[250]+10), arrowprops=dict(facecolor="black", headwidth=4, width=2, headlength=4), horizontalalignment='right',verticalalignment='top')

ax.plot(randn_cumsum)
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

np_random = np.random.randn(50)

ax1.hist(np_random, bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(50), np.arange(50)+3*np_random)

plt.show()
```

```python
np_randn = np.random.randn(50)

fig, axes = plt.subplots(2,3)

axes[0,0].hist(np_randn, bins=20, color='k', alpha=0.3)

plt.show()
```

```python
#for문 사용하기
fig, axes = plt.subplots(2,2)

for i in range(2):
  for j in range(2):
    axes[i,j].hist(np.random.randn(500),bins=50, color="k", alpha=0.5)

fig.subplots_adjust(wspace=1, hspace=1) #그래프 간 간격 조절(float형)
plt.show()
```

```python
#그래프 겹치기
plt.plot(randn_cumsum, "ko--") 
plt.plot(randn_cumsum, linestyle="--", color="k", marker="o")#라인스타일 --, 검정색, 마커ㅇ
plt.plot(randn_cumsum, "ko--", np_randn, "r-")
plt.show()
```

```python
#판다스에 바로 그래프 그리기

ss = pd.Series(np.random.randn(10).cumsum(),
               index=np.arange(0,100,10))
ax = ss.plot()
plt.show()
```
