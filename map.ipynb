```python
# CSV 파일 열기
import pandas as pd

apt_info = pd.read_csv("aptinfo.csv",encoding='cp949')
print(apt_info.head(5))
```

```python
import folium
```

```python
map = folium.Map(location=[37.519671, 127.055517],zoom_start=5)
map
```

```python
for i in range(len(apt_info)):
  marker01 = folium.Marker([apt_info.loc[i]['x'], apt_info.loc[i]['y']],
                           icon=folium.Icon(color='blue'))
  marker01.add_to(map)
```
